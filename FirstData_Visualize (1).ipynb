```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv('/content/drive/MyDrive/PajiWife/1.csv')

# Display the first five records
df.head()
```

```python

# List the column names
print(df.columns.tolist())
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
# Plotting top service types
top_service_types = df['Service types'].value_counts().head(6)
top_service_types.plot(kind='bar', color='skyblue')
plt.title('Top Service Types')
plt.xlabel('Service Type')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

```python
# Convert 'Report publication date' to datetime
df['Report publication date'] = pd.to_datetime(df['Report publication date'])

# Plotting reports over time
df['Report publication date'].value_counts().sort_index().plot(kind='line', marker='o')
plt.title('Reports Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Reports')
plt.show()
```

```python
# Plotting distribution of phone numbers
df['Phone number'].hist(bins=20, color='lightgreen')
plt.title('Distribution of Phone Numbers')
plt.xlabel('Phone Number')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plotting top local authorities
top_local_authorities = df['Local authority'].value_counts().head(10)
top_local_authorities.plot(kind='bar', color='skyblue')
plt.title('Top Local Authorities with Providers')
plt.xlabel('Local Authority')
plt.ylabel('Number of Providers')
plt.xticks(rotation=45)
plt.show()
```

```python
# Extract top 10 service types
top_service_types = df['Service types'].value_counts().head(5)

# Plotting pie chart of top 10 service types distribution
plt.figure(figsize=(8, 8))
plt.pie(top_service_types, labels=top_service_types.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 10 Service Types Distribution')
plt.show()
```

```python
# Plotting histogram of provider count by local authority
plt.figure(figsize=(10, 6))
df['Local authority'].value_counts().plot(kind='hist', bins=20, color='skyblue')
plt.title('Provider Count by Local Authority')
plt.xlabel('Number of Providers')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plotting bar chart of provider count by region
plt.figure(figsize=(10, 6))
df['Region'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Provider Count by Region')
plt.xlabel('Region')
plt.ylabel('Number of Providers')
plt.xticks(rotation=45)
plt.show()
```
